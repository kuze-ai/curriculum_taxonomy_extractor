```python
import rdflib

g = rdflib.Graph()
g.parse("../data/raw/main.ttl", format="ttl")
```

```python
## Get all the types of nodes in the graph
types = set()
for s, p, o in g:
    types.add(type(o))
```

```python
## Get the first 10 nodes randomly
import random
random.seed(42)
nodes = random.sample(list(g), 10)
## show the node names, titles and types
for n in nodes:
    print(n, g.value(n), type(n))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Convert RDF graph to a NetworkX graph
G = nx.DiGraph()
for index, (subj, pred, obj) in enumerate(g):
    if index >= 2:  # Limit to the first 100 edges
        break
    G.add_edge(subj, obj, label=pred)

# Draw the graph
pos = nx.spring_layout(G)  # Layout algorithm for positioning nodes
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='lightblue', linewidths=0.25, font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): G.edges[u, v]['label'] for u, v in G.edges})
plt.show()
```
